```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tech_layoffs_hiring_trends_elite_v2.csv')

print(df.info())
print(df.head())
```

```python
df
```

```python
sns.set_theme(style = "whitegrid")
```

```python
#Total Layoffs By Industry

plt.clf()
df_industry = df.groupby("industry")["layoffs_count"].sum().reset_index().sort_values(by = "layoffs_count",ascending = False)
sns.barplot(data= df_industry, x = "layoffs_count", y = "industry", hue = "industry", palette = "plasma", legend = False)
plt.title("Total Layoffs By Industry")
plt.xlabel("Total Layoffs")
plt.ylabel("Industry")
plt.tight_layout()
plt.savefig("layoffs_by_industry.png")
```

```python
df_industry
```

```python
#Reason for layoffs

plt.clf()
df_reason = df["reason_for_layoffs"].value_counts().reset_index()
plt.pie(df_reason["count"], labels = df_reason["reason_for_layoffs"], autopct = "%1.1f%%", startangle = 140, colors = sns.color_palette("pastel"))
plt.title("Main Reason For Layoffs")
plt.tight_layout()
plt.savefig("reason_for_layoffs.png")
```

```python
#Top Hiring Roles

plt.clf()
df_roles = df["top_hiring_role"].value_counts().reset_index().head(10)
sns.barplot(data = df_roles, x = "count", y = "top_hiring_role", hue = "top_hiring_role", palette = "mako", legend = False)
plt.title("Top 10 Hiring Roles")
plt.xlabel("count")
plt.ylabel("top_hiring_role")
plt.tight_layout()
plt.savefig("top_hiring_roles.png")

```

## 📝 Project Overview
This project analyzes a comprehensive dataset of 12,000 tech industry records to decode the underlying patterns driving the recent wave of layoffs and shifting hiring trends. Beyond surface-level numbers, the analysis uncovers how technological disruption and economic market conditions are forcing companies to restructure their workforce.

---

## 🔍 Key Insights & Data-Driven Findings

### 📉 Step 1: Industry-Wise Impact Analysis
* **Social Media & AI Lead Layoffs:** Out of all sectors in the dataset, **Social Media** recorded the highest absolute number of job cuts with over **9.01 million layoffs**, closely followed by the **AI industry** with **8.75 million layoffs**.
* **The Structural Pivot:** Although the AI and E-Commerce sectors saw massive workforce reductions, they are not shrinking. Instead, companies in these domains are aggressively shifting capital away from traditional departments to fund advanced engineering teams.

![Industry Layoffs](1_layoffs_by_industry.png)

---

### 🧩 Step 2: Primary Drivers of Layoffs
* **The Rise of Automation:** **AI Automation** emerged as the single largest reason for workforce displacement, accounting for **20.3%** of all layoffs in the dataset. This proves that automation is directly restructuring the modern tech workforce.
* **Economic & Strategic Pressures:** Strategic recalibrations like **Overhiring Correction (20.3%)** and **Restructuring (20.0%)** closely follow, showing that companies are actively trimming corporate bloat built up during previous hyper-growth years.

![Reasons for Layoffs](2_reasons_for_layoffs.png)

---

### 🚀 Step 3: Shifting Talents & High-Demand Roles
* **The AI Talent Premium:** There is a massive hiring boom for specialized AI roles. **ML Engineers (2,190 open role instances)** and **Data Scientists (2,119 open role instances)** heavily dominate the market demand.
* **Legacy vs. Next-Gen:** Traditional roles have taken a back seat, whereas infrastructure and intelligence roles like ML, Data Science, Frontend, and DevOps represent the resilient core of modern tech hiring trends.

![Top Hiring Roles](3_top_hiring_roles.png)

---

## 💡 Strategic Takeaways & Conclusion
1. **Talent Re-skilling is Mandatory:** The clear correlation between AI automation and tech layoffs highlights that professionals who do not adapt to AI tools face a higher displacement risk.
2. **Market Resilience:** Despite heavy layoffs during tough economic phases, the high volume of open roles for data and AI engineers proves that the tech ecosystem is structurally healthy but aggressively recalibrating its priorities.
